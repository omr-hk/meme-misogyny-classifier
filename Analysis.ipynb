```python
path = '/content/drive/MyDrive/project_folder/meme-misogyny-classifier/TRAINING/'
```

```python
l = []
for fn in train_labels_df['file_name']:
  print(fn)
  img = Image.open(f'{path}{fn}')
  l.append((img.width, img.height))
```

```python
max_height = 0
max_width = 0
for w, h in l:
  if w > max_width:
    max_width = w
  if h > max_height:
    max_height = h
print('max width: ', max_width)
print('max height: ', max_height)
```
